```python
import pandas as pd
```

```python
df = pd.read_excel("ventes.xlsx", header=0)
df
```

```python
df.columns
```

```python
df["Chiffre d'Affaires"] = df["Quantité"] * df["Prix  Unitaire(€)"]
df
```

```python
df.info()
```

```python
total = 0
for v in df["Chiffre d'Affaires"]:
  total += v
print(total)
```

```python
total_ca = df["Chiffre d'Affaires"].sum()
print(str(total_ca) + " €")
```

```python
produit_max = df.loc[df["Chiffre d'Affaires"].idxmax(), "Produit"]
print(produit_max)
```
